```python
import pickle
```

```python
file_path = "../news_crawling/data/100/2024-09-19_14.pkl"
```

```python
with open(file_path, 'rb') as f:
	articles = pickle.load(f)
```

```python
all_contents = ''

for article in articles:
    content = article['content']
    all_contents += content

# print(all_contents)
```

```python
import matplotlib.pyplot as plt

from wordcloud import WordCloud
%matplotlib inline
```

```python
wordcloud = WordCloud(font_path='../data/NanumGothic.ttf').generate(''.join(all_contents))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
word_tokens = okt.morphs(all_contents)
```

```python
okt.morphs(all_contents, stem=True)
```

```python
stopwords_list = []

stopwords = open('../data/korean_stopwords.txt', 'r', encoding='utf-8')

for line in stopwords.readlines():
    stopwords_list.append(line.rstrip())

stopwords.close()

print(stopwords_list)
    
```

```python
remove_stopwords_result = [word for word in word_tokens if not word in stopwords_list]
print(remove_stopwords_result)
print(len(remove_stopwords_result))
```
